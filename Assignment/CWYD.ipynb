```python
# Function to parse the query using LLM
def parse_query(query):
    messages = [
        {"role": "system", "content": "You are a helpful assistant who extracts and formats feature values from queries about house prices."},
        {"role": "user", "content": f"Extract the values for the following features from the query: {', '.join(feature_names)}. Query: '{query}'"}
    ]
    
    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",  # Use the specific model you are working with
        messages=messages,
        max_tokens=150
    )
    
    result = response.choices[0].message['content'].strip()
    print(f"Extracted Values: {result}")  # Debugging statement
    return result

def clean_extracted_values(extracted_values):
    features = {}
    for line in extracted_values.split('\n'):
        if ':' in line:
            key, value = line.split(':', 1)
            key = key.strip().lstrip('- ')  # Remove any leading dashes or spaces
            features[key] = value.strip()
    
    print(f"Cleaned Features for Filtering: {features}")  # Debugging statement
    return features

def filter_data(features):
    filtered_data = data.copy()
    for feature, value in features.items():
        if feature in filtered_data.columns:
            filtered_data = filtered_data[filtered_data[feature] == value]
        else:
            print(f"Feature '{feature}' not found in dataset columns.")
    
    print(f"Filtered Data Shape: {filtered_data.shape}")  # Debugging statement
    return filtered_data

def estimate_sales_price(filtered_data):
    if 'SalePrice' not in filtered_data.columns:
        return "The 'SalePrice' column is missing in the filtered data."
    
    if not filtered_data.empty:
        average_price = filtered_data['SalePrice'].mean()
        return average_price
    else:
        return "No data available for the given features."

def handle_query(query):
    # Extract feature values using LLM
    extracted_values = parse_query(query)
    
    # Clean and parse feature values
    features = clean_extracted_values(extracted_values)
    
    # Filter the data based on the cleaned feature values
    filtered_data = filter_data(features)
    
    # Estimate the SalesPrice
    estimated_price = estimate_sales_price(filtered_data)
    
    # Print the result
    print(f"Estimated SalesPrice: {estimated_price}")

# Example usage
user_query = "What are the expected SalesPrice for houses with an OverallQual of 9?"
handle_query(user_query)
```
