```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ChickenRiceDatasetNew.csv')
df.head()
```

```python
# Make a heatmap for the float and int columns only
import seaborn as sns

# Select only float and int columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the heatmap
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
```

```python
import matplotlib.pyplot as plt

# Histogram
numeric_df.hist(bins=15, figsize=(15, 10))
plt.suptitle('Histograms of Numeric Columns')
plt.show()

# Box Plot
numeric_df.plot(kind='box', subplots=True, layout=(2, 3), figsize=(15, 10), sharex=False, sharey=False)
plt.suptitle('Box Plots of Numeric Columns')
plt.show()

# Scatter Plot
# Assuming 'col1' and 'col2' are two numeric columns in your DataFrame
plt.figure(figsize=(10, 6))
plt.scatter(numeric_df['col1'], numeric_df['col2'])
plt.title('Scatter Plot between col1 and col2')
plt.xlabel('col1')
plt.ylabel('col2')
plt.show()

# Line Plot
# Assuming 'time' is a time series column and 'value' is a numeric column
plt.figure(figsize=(10, 6))
plt.plot(numeric_df['time'], numeric_df['value'])
plt.title('Line Plot of value over time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

# Pair Plot using seaborn for better visualization
import seaborn as sns
sns.pairplot(numeric_df)
plt.suptitle('Pair Plot of Numeric Columns', y=1.02)
plt.show()
```
